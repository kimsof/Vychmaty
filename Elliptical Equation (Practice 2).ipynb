### Решить задачу:
### $ \left\{\begin{matrix}
\Delta u = 1, x=y=0.5\\
\Delta u = 0, x, y \neq 0.5\\ 
0\leq x\leq 1, 0 \leq y \leq 1 \\
u(0, y) = 1, u(1, y) = 1\\ 
u(x, 0) = 1, u(x, 1) = 1\\ 
\end{matrix}\right. $

### Разностная схема:
# $ \frac{u_{m-1,l} - 2u_{m,l} + u_{m+1,l}}{h_{x}^{2}} + \frac{u_{m,l-1} - 2u_{m,l} + u_{m,l+1}}{h_{y}^{2}} = f $

#### 1) Ввод параметров задачи:

```python
import numpy as np
import pandas as ps
import matplotlib.pyplot as plt

def table(array, x, y):                 # для вывода таблицы с ответом
    return ps.DataFrame(array, y, x)

def plot(u, x, y):                      # для построения графиков
    fig = plt.figure()
    ax = fig.add_subplot(1, 1, 1, projection='3d')
    x, y = np.meshgrid(x, y)
    #u = function(p)
    ax.plot_surface(x, y, u)
    plt.show()

class Parameters(object):               # параметры задачи
    pass
p = Parameters() 
p.x0 = 0                                # границы по х
p.xM = 2          
p.y0 = 0                                # границы по y
p.yL = 1          
p.h = 0.1                               # шаг по x
p.f = 0                                 # неоднородность
p.M = int((p.xM - p.x0)/p.h)
p.L = int((p.yL - p.y0)/p.h)
p.x = np.linspace(p.x0, p.xM, p.M + 1)   
p.y = np.linspace(p.y0, p.yL, p.L + 1)

# граничные функции
def phi_0(y):
    return 2
def phi_M(y):
    return 2
def psi_0(x):
    return 2
def psi_L(x):
    return 2
```

#### 2) Сборка матриц А, u в начальном приближении, F для системы Au = F:

```python
def AuF(p):
    
    A = np.zeros(((p.M + 1)*(p.L + 1), (p.M + 1)*(p.L + 1)))

    # u в начальном приближении = 0 везде, кроме границ
    u = np.zeros(((p.L + 1),(p.M + 1)))
    u[0] = psi_0(p.x)
    u[p.L] = psi_L(p.x)
    u.T[0] = phi_0(p.y)
    u.T[p.M] = phi_M(p.y)
    u = u.reshape((p.M + 1)*(p.L + 1))

    # F для u на границах (ниже дозаполнится)
    F = np.zeros(((p.L + 1),(p.M + 1)))
    F[0] = psi_0(p.x)
    F[p.L] = psi_L(p.x)
    F.T[0] = phi_0(p.y)
    F.T[p.M] = phi_M(p.y)
    F = F.reshape((p.M + 1)*(p.L + 1))

    m = list(p.x).index(0.5) # номера индексов для случая x=y=0.5
    l = list(p.y).index(0.5)
    I = m + l*(p.M + 1)

    # для граничных u при y[0] А имеет только глав. диагональ = 1
    for i in range(0, (p.M + 1) + 1):
        A[i][i] = 1 


    for i in range(p.M + 2, (p.M - 1 + (p.L - 1)*(p.M + 1)) + 1):
        # для граничных u при x[0] и x[M] А имеет только глав. диагональ = 1
        if i % (p.M + 1) == 0 or (i - p.M) % (p.M + 1) == 0:
            A[i][i] = 1

        # для неграничных u
        else:
            A[i][i] = -4/p.h**2
            A[i][i - 1] = 1/p.h**2
            A[i][i + 1] = 1/p.h**2
            A[i][i - 1 - p.M] = 1/p.h**2
            A[i][i + 1 + p.M] = 1/p.h**2
            if i == I:
                F[i] = 1
            else: 
                F[i] = 0

    # для граничных u при y[L] А имеет только глав. диагональ = 1
    for i in range(p.L*(p.M + 1) - 1, (p.M + p.L*(p.M + 1)) + 1):
        A[i][i] = 1 
        
    #A = ps.DataFrame(A)
    #A.to_excel("matrix_A.xlsx")

    return A, u, F
```

#### 3) Решение системы методом linalg.solve:

```python
A = AuF(p)[0]
F = AuF(p)[2]
u_solve = np.linalg.solve(A, F)   
u_solve = u_solve.reshape(((p.L + 1), (p.M + 1)))

plot(u_solve, p.x, p.y)
```

#### 4) Решение системы МПИ:

```python
A = AuF(p)[0]
u = AuF(p)[1]
F = AuF(p)[2]
tau = 0.1
N = 100    # число итераций
for n in range(N + 1):
    B = np.eye((p.M + 1)*(p.L + 1)) - tau*A
    u_new = np.dot(B, u) + tau*F
    u = u_new
u_mpi = u_new.reshape(((p.L + 1),(p.M + 1)))
plot(u_mpi, p.x, p.y)
```

#### 5) Применим предобуславливатель Гаусса-Зейделя:

```python
A = AuF(p)[0]
u = AuF(p)[1]
F = AuF(p)[2]

U = np.triu(A, 1) 
L = np.tril(A, -1)
D = np.diag(np.diag(A))
B = np.linalg.inv(L + D)

N = 1000  # число итераций
for n in range(N + 1):
    u_new = - np.dot(np.dot(B, U), u) + np.dot(B, F)
    u = u_new
    
u_GZ = u.reshape(((p.L + 1),(p.M + 1)))
plot(u_GZ, p.x, p.y)
```
