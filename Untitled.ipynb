### Решить задачу:
### $ \left\{\begin{matrix}
\Delta u = 0\\ 
0\leq x\leq 1, 0 \leq y \leq 1 \\
u(0, y) = 1, u(1, y) = 1\\ 
u(x, 0) = 1, u(x, 1) = 1\\ 
\end{matrix}\right. $

### Разностная схема:
# $ \frac{u_{m-1,l} - 2u_{m,l} + u_{m+1,l}}{h_{x}^{2}} + \frac{u_{m,l-1} - 2u_{m,l} + u_{m,l+1}}{h_{y}^{2}} = f $

```python
import numpy as np
import pandas as ps
import matplotlib.pyplot as plt

def table(array, x, y):                 # для вывода таблицы с ответом
    return ps.DataFrame(array, y, x)

def plot(u, x, y):                      # для построения графиков
    fig = plt.figure()
    ax = fig.add_subplot(1, 1, 1, projection='3d')
    x, y = np.meshgrid(x, y)
    #u = function(p)
    ax.plot_surface(x, y, u)
    plt.show()

class Parameters(object):               # параметры задачи
    pass
p = Parameters() 
p.x0 = 0                                # границы по х
p.xM = 2          
p.y0 = 0                               # границы по y
p.yL = 1          
p.h = 0.1                               # шаг по x
p.N = 15
p.tau = 1
p.f = 0                                 # неоднородность
p.M = int((p.xM - p.x0)/p.h)
p.L = int((p.yL - p.y0)/p.h)
p.x = np.linspace(p.x0, p.xM, p.M + 1)   
p.y = np.linspace(p.y0, p.yL, p.L + 1)

# граничные функции
def phi_0(y):
    return 2
def phi_M(y):
    return 2
def psi_0(x):
    return 2
def psi_L(x):
    return 2
```

```python
# div grad u = f
# МПИ
def function_mpi(p):

    # u в начальном приближении = 0 везде, кроме границ
    u = np.zeros(((p.L + 1),(p.M + 1)))
    u[0] = psi_0(p.x)
    u[p.L] = psi_L(p.x)
    u.T[0] = phi_0(p.y)
    u.T[p.M] = phi_M(p.y)
    
    
    # Au = F
    A = np.zeros(((p.M + 1)*(p.L + 1), (p.M + 1)*(p.L + 1)))
    F = u  # F для u на границах (ниже дозаполнится)
    
    u = u.reshape((p.M + 1)*(p.L + 1))
    F = F.reshape((p.M + 1)*(p.L + 1))
    
    # для граничных u при y[0] А имеет только глав. диагональ = 1
    for i in range(0, (p.M + 1) + 1):
        A[i][i] = 1 
        
    
    for i in range(p.M + 2, (p.M - 1 + (p.L - 1)*(p.M + 1)) + 1):
        
        # для граничных u при x[0] и x[M] А имеет только глав. диагональ = 1
        if i % (p.M + 1) == 0 or (i - p.M) % (p.M + 1) == 0:
            A[i][i] = 1
            
        # для неграничных u
        else:
            A[i][i] = -4/p.h**2
            A[i][i - 1] = 1/p.h**2
            A[i][i + 1] = 1/p.h**2
            A[i][i - 1 - p.M] = 1/p.h**2
            A[i][i + 1 + p.M] = 1/p.h**2
            F[i] = p.f
            
    # для граничных u при y[L] А имеет только глав. диагональ = 1
    for i in range(p.L*(p.M + 1) - 1, (p.M + p.L*(p.M + 1)) + 1):
        A[i][i] = 1 
    
    # итерационный процесс
    for n in range(p.N + 1):
        u_new = u - p.tau*np.dot(A,u) + p.tau*F
        u_new = u_new.reshape(((p.L + 1),(p.M + 1)))
        u_new[0] = psi_0(p.x)
        u_new[p.L] = psi_L(p.x)
        u_new.T[0] = phi_0(p.y)
        u_new.T[p.M] = phi_M(p.y)
        u_new = u_new.reshape((p.M + 1)*(p.L + 1))
        u = u_new
        
    u = u.reshape(((p.L + 1),(p.M + 1)))
    return u

u_mpi = function_mpi(p)
```

```python
table(u_mpi, p.x, p.y)
```

```python
# численное решение
plot(u_mpi, p.x, p.y)
```

```python
# div grad u = f
# linalg.solve
def function_solve(p):
    
    # Au = F
    A = np.zeros(((p.M + 1)*(p.L + 1), (p.M + 1)*(p.L + 1)))
    
    # F для u на границах (ниже дозаполнится)
    F = np.zeros(((p.L + 1),(p.M + 1)))
    F[0] = psi_0(p.x)
    F[p.L] = psi_L(p.x)
    F.T[0] = phi_0(p.y)
    F.T[p.M] = phi_M(p.y)
    F = F.reshape((p.M + 1)*(p.L + 1))
    
    m = list(p.x).index(0.5) # номера индексов для случая x=y=0.5
    l = list(p.y).index(0.5)
    I = m + l*(p.M + 1)
    
    # для граничных u при y[0] А имеет только глав. диагональ = 1
    for i in range(0, (p.M + 1) + 1):
        A[i][i] = 1 
    
    
    for i in range(p.M + 2, (p.M - 1 + (p.L - 1)*(p.M + 1)) + 1):
        # для граничных u при x[0] и x[M] А имеет только глав. диагональ = 1
        if i % (p.M + 1) == 0 or (i - p.M) % (p.M + 1) == 0:
            A[i][i] = 1
            
        # для неграничных u
        else:
            A[i][i] = -4/p.h**2
            A[i][i - 1] = 1/p.h**2
            A[i][i + 1] = 1/p.h**2
            A[i][i - 1 - p.M] = 1/p.h**2
            A[i][i + 1 + p.M] = 1/p.h**2
            if i == I:
                F[i] = 1
            else: 
                F[i] = 0
            
    # для граничных u при y[L] А имеет только глав. диагональ = 1
    for i in range(p.L*(p.M + 1) - 1, (p.M + p.L*(p.M + 1)) + 1):
        A[i][i] = 1 
    
    u =np.linalg.solve(A, F)
        
    u = u.reshape(((p.L + 1), (p.M + 1)))
    
    return u

u_solve = function_solve(p)
```

```python
table(u_solve, p.x, p.y)
```

```python
# численное решение
plot(u_solve, p.x, p.y)
```
