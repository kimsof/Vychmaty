```python
import numpy as np
x = [-0.1, 0.1, 0.3, 0.5, 0.7, 0.9, 1.1, 1.3, 1.5, 1.7]
f = [0.5487, 0.5487, 0.6137, 0.7317, 0.8727, 0.9820, 0.9780, 0.7712, 0.3153, -0.3138]
h = 0.2

u = np.zeros(10)
for i in range(1,10):
    u[i] = (f[i] - f[i-1])/h
    
p = np.zeros(10)
p[1] = -h/(4*h)
for i in range(3,10):
    p[i-1] = -h/(4*h + h*p[i-2])
    
q = np.zeros(10)
q[1] = 3*(u[2]-u[1])/(4*h)
for i in range(3,10):
    q[i-1] = (3*u[i] - 3*u[i-1] - h*q[i-2])/(4*h + h*q[i-2])
    
c = np.zeros(10)
for i in reversed(range(2,10)):
    c[-(i-1)] = p[-(i-1)] + q[-(i-1)]
    
b = np.zeros(10)
for i in range(1,10):
    b[i] = u[i] + (2*c[i]*h + h*c[i-1])/3
    
    
d = np.zeros(10)
for i in range(1,10):
    d[i] = (c[i] - c[i-1])/(3*h)

a = np.zeros(10)
for i in range(1,10):
    a[i] = f[i]
    
def interp0late(k):
    S = np.zeros(10)
    v = 0
    w = 0
    for i in range(1,10):
        S[i] = a[i] + b[i]*(k-x[i]) + c[i]*(k-x[i])**2 + d[i]*(k-x[i])**3
    while v<10:
        if (x[v] <= k) & (k < x[v+1]) :
            w = S[v]
            v=11
        else:
            v+=1
    return w
```

```python
import pandas as ps
mas = np.zeros((8,10))
mas[0,:] = u[:]
mas[1,:] = p[:]
mas[2,:] = q[:]
mas[3,:] = a[:]
mas[4,:] = b[:]
mas[5,:] = c[:]
mas[6,:] = d[:]
tab = ps.DataFrame(data=mas, index=np.arange(0,8), columns=np.arange(0,10))
tab.index = ['u','p','q','a','b','c','d','comppa']
```
