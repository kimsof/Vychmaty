```python
def f(x):
    return x * np.exp(-x * x)

x_max = 1/2**(1/2)
f_max = f(x_max)
epsilon_x = 0.001
epsilon_f = (np.exp(-x_max*x_max) * (1 - 2*x_max*x_max))**(1/2)
x0_left = 0
x0_right = 2

while np.fabs(f(x0_left) - f_max / 2) > epsilon_f:
    x0_left = np.exp(x0_left**2) * f_max / 2
    
while np.fabs(f(x0_right) - f_max / 2) > epsilon_f:
    x0_right = (np.log(2 * x0_right/ f_max))**(1/2)

width = round(x0_right, 3) - round(x0_left, 3)

print("ширина полувысоты при  ε = 0.001: " + str(width))
```
